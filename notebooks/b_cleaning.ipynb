Our unique() check reveals that total_sqft is a StringArray containing values like '1020 - 1130'. Since a machine learning model only understands single numbers, we need to convert these ranges to their average.

````python
# Let's check total_sqft is float or not
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
````

````python
df[~df['total_sqft'].apply(is_float)].head(10)
````

````python
# as we can see the there is massy data let's convert into number
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
````

<h3 style="color:white">Outlier Removal Using Business Logic</h3>

As a data scientist when you have a conversation with your business manager (who has expertise in real estate), he will tell you that normally square ft per bedroom is 300 (i.e. 2 bhk apartment is minimum 600 sqft. If you have for example 400 sqft apartment with 2 bhk than that seems suspicious and can be removed as an outlier. We will remove such outliers by keeping our minimum thresold per bhk to be 300 sqft

````python
df[df.total_sqft/df.bhk<300].head()
````

Check above data points. We have 6 bhk apartment with 1020 sqft. Another one is 8 bhk and total sqft is 600. These are clear data errors that can be removed safely

````python
df[df.total_sqft/df.bhk<300].value_counts().sum()
````

````python
# Now we can Remove those outliers where total_sqft per bhk is less than 300 sqft
df[~(df.total_sqft / df.bhk < 300)]
````

**Completed Steps**

**Imputation:** We successfully filled missing **bath** values using the median.

**Feature Extraction:** We created the numerical **bhk** column from the **size** string and dropped the original size column.

**Structural Conversion:** We converted total_sqft from messy strings and ranges into clean floating-point numbers.

**Outlier Identification:** We identified illogical data points (e.g., 43 bedrooms in 2400 sqft).

### This is for ```src/components/b_cleaning.py```

````python
import os
import sys
import pandas as pd
import numpy as np
import logging

logger = logging.getLogger(__name__)


class DataCleaning:
    def __init__(self):
        pass

    def convert_sqft_to_num(self, x):
        """Verified Logic: Converts ranges to average and handles units."""
        try:
            tokens = str(x).split('-')
            if len(tokens) == 2:
                return (float(tokens[0]) + float(tokens[1])) / 2
            return float(x)
        except Exception:
            return None

    def start_data_cleaning(self, df: pd.DataFrame) -> pd.DataFrame:
        """
        Finalized cleaning function based on verified notebook steps.
        """
        try:
            logger.info("Starting production data cleaning process...")
            df = df.copy()

            # 1. Drop Duplicates
            df = df.drop_duplicates()

            # 2. Handle Missing Values for critical features
            df = df.dropna(subset=['size', 'location'])

            # 3. Impute 'bath' with median
            df['bath'] = df['bath'].fillna(df['bath'].median())

            # 4. Create 'bhk' and drop 'size'
            df['bhk'] = df['size'].apply(lambda x: int(x.split(' ')[0]))
            df = df.drop(['size'], axis=1)

            # 5. Clean 'total_sqft'
            df['total_sqft'] = df['total_sqft'].apply(self.convert_sqft_to_num)
            
            # 6. Drop those 46 new Nulls in total_sqft
            df = df.dropna(subset=['total_sqft'])

            # 7. Business Logic Outlier Removal
            # Removing houses where sqft per room is less than 300
            df = df[~(df.total_sqft / df.bhk < 300)]
            
            logger.info(f"Cleaning complete. Resulting shape: {df.shape}")
            return df

        except Exception as e:
            logger.error("Error occurred in Data Cleaning component")
            raise 
````
